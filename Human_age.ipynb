# ***DATA INSPECTION***

```python
# prompt: # Get the number of samples and features
#First convert the 'Gender' column to numerical values using LabelEncoder
encoder = LabelEncoder()
df['Gender'] = encoder.fit_transform(df['Gender'])

# Now you can square the 'Gender' column
df['Gender_squared'] = df['Gender'] ** 2
print(df.info())

n_samples = df.shape[0]
n_features = df.shape[1]
print(df.describe())
# prompt: create a feature

# Assuming 'Age' is a column in your dataframe
df['Gender_squared'] = df['Gender'] ** 2

print("Number of samples:", n_samples)
print("Number of features:", n_features)
print(df.isnull().sum())
```

# ***DATA PROCESSING***

```python
# Check for missing values in the dataset
missing_values = df.isnull().sum()
print(missing_values)
```

```python
# Impute missing Age values with the median age
df['Gender'].fillna(df['Gender'].median(), inplace=True)
```

```python
numerical_features = ['Gender', 'Height (cm)']
print("Numerical Features Before Scaling:")
print(df[numerical_features].head())
```

```python
# Apply Standardization
scaler = StandardScaler()
df[numerical_features] = scaler.fit_transform(df[numerical_features])

print("Numerical Features After Standardization:")
print(df[numerical_features].head())
```

```python
# Apply Min-Max Scaling
min_max_scaler = MinMaxScaler()
df[numerical_features] = min_max_scaler.fit_transform(df[numerical_features])

print("Numerical Features After Min-Max Scaling:")
print(df[numerical_features].head())
```

```python
# Initialize the LabelEncoder
label_encoder = LabelEncoder()

# Apply Label Encoding to the 'Pclass' feature
df['Pclass'] = label_encoder.fit_transform(df['Pclass'])

# Print the first few rows of the encoded feature
print("Pclass Feature After Label Encoding:")
print(df[['Pclass']].head())

# Print mapping of labels to original values
print("Label Encoding Mapping:")
for i, category in enumerate(label_encoder.classes_):
    print(f"{category}: {i}")
```
